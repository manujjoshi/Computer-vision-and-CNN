```python
img  = cv2.imread(r'/content/drive/MyDrive/DL-latest/Resources/Photos/park.jpg')
cv2_imshow(img)
cv2.waitKey(0)
```

```python
No_of_pixel = img.shape[0]*img.shape[1]*img.shape[2]
print(No_of_pixel)
```

```python
gra = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gra)
cv2.waitKey(0)
```

```python
gra.shape
```

```python
blank  = np.zeros(gra.shape[:2], np.uint8)
```

```python
intersect = cv2.circle(blank, (350, 150), 100, 255, -1)
```

```python
masked = cv2.bitwise_and(gra, gra, mask = intersect)
```

```python
cv2_imshow(masked)
cv2.waitKey(0)
```

```python
sobel_x = cv2.Sobel(gra, cv2.CV_64F, 1, 0)
sobel_y = cv2.Sobel(gra, cv2.CV_64F, 0, 1)
cv2_imshow(sobel_x)
cv2_imshow( sobel_y)
combined_sobel = cv2.bitwise_or(sobel_x, sobel_y)
cv2_imshow(combined_sobel)
cv2.waitKey(0)
```

```python
alpha = 2.2
beta= 50
new_img = cv2.convertScaleAbs(img, alpha=alpha, beta = beta)
cv2_imshow(new_img)
cv2.waitKey(0)
```
