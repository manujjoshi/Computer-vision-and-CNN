# IOU - Intersection Over Union 

```python
import cv2
import pandas as pd 
from matplotlib import pyplot as plt 
```

```python
def IOU(box1, box2): 
    x1,y1,x2,y2 = box1
    x3,y3,x4,y4  = box2
    print(x1, y1,x2, y2)
    print(x3,y3,x4,y4)
    x_int_1= max(x1,x3)
    y_int_1= max(y1,y3)
    x_int_2= min(x2,x4)
    y_int_2= min(y2,y4)
    width_inter = abs(x_int_2 - x_int_1)
    height_inter = abs(y_int_2 - y_int_1)
    area_inter = width_inter * height_inter
    #Now we will calculate the union of the bounding boxes
    width_box1 = abs(x2-x1)
    height_box1 = abs(y2-y1)
    width_box2  = abs(x4-x3)
    height_box2 = abs(y4-y3)
    area_box1 = width_box1 * height_box1
    area_box2 = width_box2 * height_box2
    area_union = area_box1 + area_box2 - area_inter
    iou = area_inter / area_union
    return iou 

def draw_rect(frame, bbox, color = (0,255,0)):
    x1, y1,x2,y2 = bbox
    cv2.rectangle(frame, pt1=(x1,y1), pt2=(x2, y2), color=color, thickness=2)
    return frame

```

```python
bbox_plane_actual = [56, 13, 180, 125]
bbox_plane_predicted = [48, 9, 172, 115]
img = cv2.imread(r"C:\Users\R1007398\Downloads\Books_AIML\Innomatics\CV\CNN\Codes\Datasets\Drone\dumps-img\airplane_007.jpg")
img = draw_rect(img, bbox_plane_actual, color = (0,255,0) )
img = draw_rect(img, bbox_plane_predicted, color = (0, 0, 255) )
iou = IOU(bbox_plane_actual, bbox_plane_predicted)
img = cv2.putText(img, 'IOU:{}'.format(iou), (bbox_plane_actual[0],bbox_plane_predicted[1]), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 1, cv2.LINE_AA)
print("IOU value  = ", iou)
plt.imshow(img)
plt.show()
```
